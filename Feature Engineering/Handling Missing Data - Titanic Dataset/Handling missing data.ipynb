```python
'''
Missing values
1. Missing completely at random (no pattern whatsoever)
2. Missing at random (there is a pattern between missing values and remaining data)
3. Missing not at random (reason is specified in dataset as to why the data is missing)
Eg : In depression analysis form, missing vales means the person is highly depressed
'''
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('../input/titanicdataset-traincsv/train.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.isnull().mean()
```

```python
# Missing data Not At Random
```

```python
# If a person did not tell their age or cabin, it is because that person probably did not survive the accident.
```

```python
data['cabin_isnull']=np.where(data['Cabin'].isnull(),1,0)
```

```python
data.head()
```

```python
data.groupby(['Survived'])['cabin_isnull'].mean()
```

```python
# So, out of all those who did not survive, 87% of data is missing
```

```python
data['Age'].isnull().groupby(data['Survived']).mean()
```

```python
# So, 22% data is missing for those who did not survive, whereas 15% data is missing for those who did survive
```
